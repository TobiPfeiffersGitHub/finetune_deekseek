```python
from transformers import AutoTokenizer, AutoModelForCausalLM, pipeline
from peft import PeftModel, PeftConfig
import torch
from unsloth import FastLanguageModel, is_bfloat16_supported
```

### Microsoft Phi

```python
peft_model_name = "tobinho1234/microsoft-phi-customer-support"

config = PeftConfig.from_pretrained(peft_model_name)

base_model = AutoModelForCausalLM.from_pretrained(config.base_model_name_or_path)

model = PeftModel.from_pretrained(base_model, peft_model_name)
model.eval()

tokenizer = AutoTokenizer.from_pretrained(config.base_model_name_or_path)
```

```python
customer_support = pipeline(
    "text-generation", 
    model=model, 
    tokenizer=tokenizer, 
    device=0 if torch.cuda.is_available() else -1
    )
```

```python
inquiries = [
    "Hi, I received the wrong item in my order. What should I do?",
    "My internet connection keeps dropping every few hours. Can you help?",
    "I want to cancel my subscription but I canâ€™t find the option online.",
    "Do you offer international shipping to Canada?",
    "My payment was declined but I have enough funds. Why did this happen?"
]
```

```python
for question in inquiries:
    print(f"ğŸ“© Customer: {question}")
    response = customer_support(question, max_new_tokens=150, do_sample=True, temperature=0.7)[0]['generated_text']
    print(f"ğŸ¤– Support Bot: {response[len(question):].strip()}\n{'-'*80}")
```

### DeepSeek
since it was trained with unsloth we have to use the unsloth loader and tokenzier

```python
model_name = "tobinho1234/deep-seek-lora-customer-support"

model, tokenizer = FastLanguageModel.from_pretrained(
    model_name = model_name,
    max_seq_length = 2048,
    dtype = None,           
    load_in_4bit = True    
)

FastLanguageModel.for_inference(model) 
```

```python
def format_prompt(message):
    return f"Customer: {message}\nSupport:"

for question in inquiries:
    prompt = format_prompt(question)
    inputs = tokenizer(prompt, return_tensors="pt").to(model.device)

    outputs = model.generate(
        **inputs,
        max_new_tokens=150,
        temperature=0.7,
        do_sample=True,
        top_p=0.9,
        top_k=50,
        pad_token_id=tokenizer.eos_token_id,
    )

    response = tokenizer.decode(outputs[0], skip_special_tokens=True)
    answer = response[len(prompt):].strip()

    print(f"ğŸ“© Customer: {question}")
    print(f"ğŸ¤– Support Bot: {answer}\n{'-'*80}")
```
